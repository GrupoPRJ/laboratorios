```python
from Bio.Blast import NCBIXML
from Bio import SearchIO

def parse_prot(file, E_VALUE_THRESH):
    #file --> nome do ficheiro .xml resultante da função blast_prot(), onde estão contidos os resultados do blast. Ex. 'TCF7L2_blastprot.xml'
    #E_VALUE_THRESH --> valor máximo para o e-value

    result_handle = open(file)
    blast_record = NCBIXML.read(result_handle)
    if E_VALUE_THRESH == None:
        E_VALUE_THRESH = 0.05
    for alignment in blast_record.alignments:
        for hsp in alignment.hsps:
            if hsp.expect < E_VALUE_THRESH:
                print("Alignment")
                print("sequence:", alignment.title)
                print("length:", alignment.length)
                print("e value:", hsp.expect)
                print(hsp.query[0:50] + "...")
                print(hsp.match[0:50] + "...")
                print(hsp.sbjct[0:50] + "...")
                print()
    
    blast_qresult = SearchIO.read(file, "blast-xml")
    print(blast_qresult)
    result_handle.close()

#Irá imprimir informações relativas ao blast realizado, contidas no ficheiro .xml
```

```python
parse_prot('TCF7L2_blastprot.xml',E_VALUE_THRESH=0.05)
```
