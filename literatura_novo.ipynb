```python
from Bio import Entrez
import pandas as pd

Entrez.email= 'pg49846@alunos.uminho.pt'

handle=Entrez.esearch(db ='pubmed', term='SLC30A8', retmax= 10)  
records=Entrez.read(handle)   
handle.close()

list_id = records['IdList']
```

```python
handle = Entrez.esummary(db='pubmed', id=list_id)
records=Entrez.parse(handle)

article_data = dict()
article_data_keys = ['Pubmed ID', 'Title', 'Source', 'Authors', 'DOI']

for x in article_data_keys:
    article_data[x] = list()

for i in records:
    article_data['Pubmed ID'].append(i['Id'])
    article_data['Title'].append(i['Title'])
    article_data['Source'].append(i['Source'])
    article_data['Authors'].append(i['AuthorList'])
    article_data['DOI'].append(i['DOI'])
    

df = pd.DataFrame(article_data)
```
